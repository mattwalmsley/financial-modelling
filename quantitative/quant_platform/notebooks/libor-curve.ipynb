```python
import matplotlib.pyplot as plt
```

```python
months = [1,2,3,6,9,12,15,18,24,48]
spot_prices = [0.804,0.997,1.30,1.41,1.61,1.82,2.03,2.17,2.30,2.69]
```

```python
plt.plot(months, spot_prices, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Time (Months)')
plt.ylabel('Spot Rate (%)')
plt.title('LIBOR Spot Rates')

# Display the plot
plt.show()
```
